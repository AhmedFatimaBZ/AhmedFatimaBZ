```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
Penguins['Species'].value_counts()
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(Penguins.corr(),annot=False)
plt.show()
```

**What are the Penguin Species? Identify the different species present in the dataset**

```python
Penguins['Species'].value_counts().plot(kind='bar',title='Penguin Species',ylabel='population')
plt.show()
```

```python
Penguins['Population'] = Penguins.groupby('Species')['Species'].transform('count')
```

```python
sns.barplot(x='Species',y='Population',hue='Species',data=Penguins)
plt.title('The Population Distribution of Penguin Species')
plt.show()
```

**Species Population per Island - Determine the population of each penguin species on different islands.**

```python
sns.barplot(x='Island',y='Population',hue='Species',data=Penguins)
plt.title('The Population Distribution of Penguin Species per Island')
plt.show()
```

**Relationship between Sex and Body Mass - Explore the relationship between the sex of penguins and their body mass.**

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Sex', y='Body mass (g)', data=Penguins, hue='Sex')
plt.title('Body Mass by Sex')
plt.xlabel('Sex')
plt.ylabel('Body Mass (g)')
plt.show()
```

```python
sns.boxplot(x='Sex',y='Body mass (g)',hue='Species',data=Penguins)
plt.title('Body Mass by Sex for Each Species')
plt.show()
```

**Clutch Completion Distribution**

```python
Penguins['Clutch completion'].value_counts().plot.pie(autopct='%1.1f%%', colors=["green","red"])
plt.title('Clutch Completion Distribution')
plt.ylabel('')
plt.show()
```

**Clutch Completion by Gender - Analyze which gender has a higher rate of completing their clutch.**

```python


plt.figure(figsize=(10, 6))


sns.countplot(x='Clutch completion', hue='Sex', data=Penguins, palette=["skyblue", "pink"],
              order=Penguins['Clutch completion'].value_counts().index,
              hue_order=Penguins['Sex'].value_counts().index,
              dodge=True)

plt.title('Clutch Completion by Gender')
plt.xlabel('Clutch Completion')
plt.ylabel('Count')
plt.show()
```

**Clutch Completion Based on Island - Investigate if clutch completion varies based on the penguin species or the island they inhabit.**

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Island', hue='Clutch completion', data=Penguins)
plt.title('Clutch Completion Distribution on Different Islands')
plt.xlabel('Island')
plt.ylabel('Count')
plt.show()
```

**Correlation between Body Mass and Culmen Length - Determine the correlation between the body mass and culmen length of penguins**

```python
sns.scatterplot(x='Body mass (g)', y='Culmen length (mm)',hue='Species', data=Penguins)
plt.title('The Relationship Between Culmen Length and Body Mass by Penguin Species')
plt.show()
```

**Correlation between Body Mass and Culmen Depth - Explore the correlation between the body mass and culmen depth of penguins**

```python
sns.scatterplot (x='Body mass (g)',y='Culmen depth (mm)',hue='Species',data=Penguins)
plt.title('The Relationship Between Culmen Depth and Body Mass by Penguin Species')
plt.show()
```

**Correlation between Body Mass and Flipper Length - Investigate the correlation between the body mass and flipper length of penguins.**

```python
sns.scatterplot(x='Body mass (g)', y='Flipper length (mm)',hue='Species', data=Penguins)
plt.title('The Relationship Between Flipper Length and Body Mass by Penguin Species')
plt.show()
```
